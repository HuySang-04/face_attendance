```python
from ultralytics import YOLO
import yaml

data_config = {
    'path': '/content/data',
    'train': 'images/train',
    'val': 'images/val',
    'nc': 2,
    'names': ['fake', 'real']
}

with open('/content/data.yaml', 'w') as f:
    yaml.dump(data_config, f, sort_keys=False)

model = YOLO('yolov8n.pt')
model.train(
    data='/content/data.yaml',
    epochs=100,
    imgsz=640,
    batch=20,
    name='face_live_detect',
    project='/content/runs',
)
```
